```python
import pandas as pd

import nltk

from nltk.sentiment.vader import SentimentIntensityAnalyzer

from nltk.corpus import stopwords

from nltk.tokenize import word_tokenize

from nltk.stem import WordNetLemmatizer

nltk.download('all')
```

```python
# Carregando a base com revisões da Amazon

df = pd.read_csv('https://raw.githubusercontent.com/pycaret/pycaret/master/datasets/amazon.csv')
```

```python
def preprocess_text(text):

    # Criando tokens
    tokens = word_tokenize(text.lower())

    # Removendo parlavra irrelevantes em inglês, pois a base das revisões é em inglês
    filtered_tokens = [token for token in tokens if token not in stopwords.words('english')]

    # Lematizando os tokens
    lemmatizer = WordNetLemmatizer()

    lemmatized_tokens = [lemmatizer.lemmatize(token) for token in filtered_tokens]

    # Construindo a string que será analisada
    processed_text = ' '.join(lemmatized_tokens)

    return processed_text


df['reviewText'] = df['reviewText'].apply(preprocess_text)
```

```python
#Instanciando o analisador
analyzer = SentimentIntensityAnalyzer()


def get_sentiment(text):

    scores = analyzer.polarity_scores(text)

    sentiment = 1 if scores['pos'] > 0 else 0

    return sentiment

df['sentiment'] = df['reviewText'].apply(get_sentiment)

print("Resultado da análise:")
print(df)
```

```python
df_grouped = df.groupby("sentiment")
print(df_grouped["sentiment"].count())
```
